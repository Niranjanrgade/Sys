```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 7, 9, 11])

# create the model with at least one layer
model = Sequential([
    # Dense is the layer having 1 input and 1 output
    Dense(units=1, input_shape=[1])
])

# optimizer will optimize the weights using error (c) value
# loss will be used to calculate the error (c) value
model.compile(optimizer="sgd", loss="mean_squared_error")

model.fit(x, y, epochs=1500)
model.predict([[10]])
```
