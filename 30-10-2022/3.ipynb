```python
ca_mode = df['ca'].mode()[0]
df['ca'].fillna(ca_mode, inplace=True)
```

```python
thal_mode = df['thal'].mode()[0]
df['thal'].fillna(thal_mode, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

sex_encoder = LabelEncoder()
df['sex'] = sex_encoder.fit_transform(df['sex'])

c_p_encoder = LabelEncoder()
df['chest_pain'] = c_p_encoder.fit_transform(df['chest_pain'])

r_ecg_encoder = LabelEncoder()
df['rest_ecg'] = r_ecg_encoder.fit_transform(df['rest_ecg'])

thal_encoder = LabelEncoder()
df['thal'] = thal_encoder.fit_transform(df['thal'])
```
