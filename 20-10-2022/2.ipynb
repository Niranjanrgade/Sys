```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
x = df[['YearsExperience']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x))
```
